# PyPDFLoader

가장 많이 쓰이는 PDF Loader입니다. page 단위로 데이터를 로드합니다. 예를 들어서 27페이지의 PDF면 27개의 문서로 로드합니다. (일반적으로 그렇다는 것이고, 항상 보장되지는 않습니다. 특정 페이지에 그림만 있고 내용만 없다거나 하면 해당 페이지는 건너뜁니다.)

```python
from langchain.document_loaders import PyPDFLoader
```

```python
## pdf 파일로드 하고 쪼개기
%%time
# loader = PyPDFLoader('https://wdr.ubion.co.kr/wowpass/img/event/gsat_170823/gsat_170823.pdf')
loader = PyPDFLoader('2023_북한인권보고서.pdf')
pages = loader.load_and_split()
print('청크의 수:', len(pages))
```

데이터의 형식은 다음을 따릅니다.  

`Document(page_content='내용', metadata={'source': 파일명, 'page': 페이지 번호})`

# fitz (이미지 추출)

```python
import io
import fitz
from PIL import Image

# docs = fitz.open('2023_북한인권보고서.pdf')
docs = fitz.open("/content/gsat_170823.pdf")
```

```python
len(docs)
```

```python
for page_num in range(len(docs)):
    page = docs.load_page(page_num)
    image_list = page.get_images(full=True)
```

```python
image_list
```

```python
# iterate over pdf pages
for page_index in range(len(docs)):
    # get the page itself
    page = docs[page_index]
    image_list = page.get_images()
    # printing number of images found in this page
    if image_list:
        print(f"[+] Found a total of {len(image_list)} images in page {page_index}")
    else:
        print("[!] No images found on page", page_index)
    for image_index, img in enumerate(page.get_images(), start=1):
        # get the XREF of the image
        xref = img[0]
        # extract the image bytes
        base_image = docs.extract_image(xref)
        image_bytes = base_image["image"]
        # get the image extension
        image_ext = base_image["ext"]
        # load it to PIL
        image = Image.open(io.BytesIO(image_bytes))
        # save it to local disk
        image.save(open(f"image{page_index+1}_{image_index}.{image_ext}", "wb"))
```
